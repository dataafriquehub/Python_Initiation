### **Liste en python**



*   Une liste est une structure de données ordonnée qui peut contenir des éléments de différents types.

*   Les éléments dans une liste sont indexés, commençant par 0 pour le premier élément.

*   Les listes sont modifiables, ce qui signifie que vous pouvez ajouter, supprimer ou modifier des éléments.




```python
# List (Liste)
person= ["ISSA", "Pauline", 15,  13.5]
print("Affichage de person :", person)
```

```python
# Accès aux éléments de la liste
premier_element = person[-3]
dernier_element = person[-1]
print("Premier element de person :", premier_element)
print("Dernier element de person :", dernier_element)
```

```python
# Modification d'un élément de la liste
person[1] = "kiwi"
print("Liste mise à jour :", person)
```

```python
# Ajout d'un élément à la fin de la liste
person.append("beninois")
print("Liste après ajout :", person)
```

```python
# Suppression d'un élément de la liste
person.remove(13.5)
print("Liste après suppression :", person)
```

### **Set (Ensemble) :**


*   Un ensemble est une collection non ordonnée d'éléments uniques.

*   Les ensembles sont utiles pour effectuer des opérations telles que l'union, l'intersection, et la différence entre des ensembles.

*   Les ensembles n'ont pas d'ordre spécifique et ne permettent pas d'indexation.

```python
# Set (Ensemble)
legumes = {"carotte", "brocoli", "poivron", "haricot"}
print("Ensemble de légumes :", legumes)
```

```python
# Ajout d'un élément à l'ensemble
legumes.add("tomate")
print("Ensemble après ajout :", legumes)
```

```python
# Suppression d'un élément de l'ensemble
legumes.remove("brocoli")
print("Ensemble après suppression :", legumes)
```

### **Tuple (Tuple) :**


*   Un tuple est une collection ordonnée et immuable d'éléments.

*   Les tuples sont similaires aux listes, mais une fois créés, leurs éléments ne peuvent pas être modifiés.

*   Les tuples sont souvent utilisés pour représenter des collections de données immuables.

```python
# Tuple (Tuple)
couleurs = ("rouge", "vert", "bleu")
print("Tuple de couleurs :", couleurs)
```

```python
# Accès aux éléments du tuple
premiere_couleur = couleurs[0]
print("Première couleur :", premiere_couleur)
```

```python
# Les tuples sont immuables, on ne peut pas modifier leurs éléments
couleurs[1] = "jaune"  # Cette ligne provoquerait une erreur
```

### **Dictionary (Dictionnaire) :**


*   Un dictionnaire est une structure de données qui associe des paires clé-valeur.

*   Chaque élément dans un dictionnaire a une clé unique et une valeur correspondante.

*   Les dictionnaires permettent un accès rapide aux valeurs en utilisant leurs clés, mais ils ne sont pas ordonnés.

```python
# Dictionary (Dictionnaire)
couleurs = ("rouge", "vert", "bleu")
personne = {"nom": "Dupont ouziel", "couleur": couleurs, "age": 30}
print("Dictionnaire de personne :", personne)
```

```python
personne["couleur"][1]
```

```python
# Accès aux valeurs du dictionnaire
nom_personne = personne["nom"]
print("Nom de la personne :", nom_personne)
```

```python
# Modification de la valeur d'une clé dans le dictionnaire
personne["age"] = 40
print("Dictionnaire mis à jour :", personne)
```

```python
# Ajout d'une nouvelle paire clé-valeur
personne["ville"] = "Paris"
print("Dictionnaire après ajout :", personne)
```

### **4- Modules et importations**

* Un module en Python est un fichier contenant des définitions et des instructions Python. Il peut inclure des variables, des fonctions...

* L'utilisation de modules permet d'organiser le code de manière modulaire et de le réutiliser facilement.



**Importation de Modules :**
