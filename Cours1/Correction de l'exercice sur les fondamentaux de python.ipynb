#### **Exercice 11 : (Système de notes)**

Demandez à l'utilisateur de saisir une note entre 0 et 100. Ensuite, affichez la mention correspondante en utilisant les catégories suivantes :

* 90-100 : Excellent
* 80-89 : Très bien
* 70-79 : Bien
* 60-69 : Assez bien
* 0-59 : Échec

```python
# Exercice 11
# Demandez à l'utilisateur de saisir une note entre 0 et 100. Ensuite, affichez la mention correspondante.
note = int(input("Entrez une note entre 0 et 100 : "))

if 90 <= note <= 100:
    print("Excellent")
elif 80 <= note <= 89:
    print("Très bien")
elif 70 <= note <= 79:
    print("Bien")
elif 60 <= note <= 69:
    print("Assez bien")
else:
    print("Échec")
```

#### **Exercice 12 : Table de multiplication**

Écrivez un programme qui demande à l'utilisateur de saisir un nombre, puis affiche la table de multiplication de ce nombre de 1 à 10.

```python
# Exercice 12
# Écrivez un programme qui demande à l'utilisateur de saisir un nombre,
# puis affiche la table de multiplication de ce nombre de 1 à 10.
nombre = int(input("Entrez un nombre : "))

for i in range(1, 11):
    produit = nombre * i
    print(f"{nombre} x {i} = {produit}")
```

#### **Exercice 13 : Somme des nombres pairs**

Écrivez un programme qui calcule la somme des nombres pairs de 1 à 100 à l'aide d'une boucle for.

```python
# Exercice 13
# Écrivez un programme qui calcule la somme des nombres pairs de 1 à 100 à l'aide d'une boucle for.
somme_pairs = 0
for nombre in range(2, 101, 2):
    somme_pairs += nombre

print("La somme des nombres pairs de 1 à 100 est :", somme_pairs)
```
