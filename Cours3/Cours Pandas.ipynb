### $$\text{Création des séries }$$

```python
serie1 = pd.Series([10, 20, 15,18])
serie1
```

## $\text{III : Chargement des données, stockage et formats de fichiers}$
### $\text{Pandas nous donne la possibilité de charger les données de plusieurs type de formats de fichiers. }
$
### $\begin{align*}
\text{read_csv} & : \text{Lire les fichiers excel de format CSV.  } \\
\text{read _excel} & : \text{Lire les fichiers excel de format XLS et XLSX .} \\
\text{read_sql} & : \text{Lire les requetes SQL } \\
\text{read_json} & : \text{Lire les fichiers au java script au format JSON.} \\
\text{read_xml} & : \text{Lire les fichiers au format XML.} \\
\text{read_html} & : \text{lire les fichiers web au format HTML.  } \\
\end{align*} \\ $
### $\text{Pandas nous donne la possibilité de stocker les structures pandas dans de plusieurs type de formats de fichiers. }
$
### $\begin{align*}
\text{to_csv} & : \text{stocker un dataframe pandas dans un fichier excel de format CSV.} \\
\text{to_excel} & : \text{stocker un dataframe pandas dans un fichier excel de format XLS et XLSX.} \\
\text{to_numpy} & : \text{stocker un dataframe pandas dans un fichier numpy } \\
\end{align*}$
