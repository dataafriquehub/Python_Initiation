## Exercice 1: Introduction aux Types de Diagrammes

**Définissez brève les types de diagrammes:**


- **Line Chart (Graphique en Ligne)** : Un graphique en ligne représente des points de données reliés par des lignes droites. Il est souvent utilisé pour illustrer les tendances ou l'évolution d'une variable au fil du temps. Chaque point sur la ligne représente la valeur d'une donnée à un moment précis, ce qui rend ce type de graphique idéal pour visualiser des séries temporelles.

- **Histogramme** : Un histogramme est un type de graphique qui représente la distribution de données numériques en formant des "bacs" ou des intervalles. Chaque bin regroupe une plage de données, et la hauteur de la barre indique la fréquence ou le nombre de données dans cet intervalle. Les histogrammes sont utiles pour comprendre la distribution, la dispersion et la centralité des données.

- **Diagramme en Barres** : Un diagramme en barres affiche des informations sous forme de barres rectangulaires horizontales ou verticales, dont la longueur est proportionnelle aux valeurs qu'elles représentent. Ce type de graphique est utilisé pour comparer des quantités ou des fréquences de différentes catégories ou groupes.

- **Diagramme de Dispersion (Scatter Plot)** : Un diagramme de dispersion présente des points de données individuels sur deux axes, montrant ainsi la relation ou la corrélation entre deux variables. Chaque point représente une observation. Ce type de graphique est utile pour détecter des tendances, des regroupements ou des anomalies dans les données.

- **Diagramme Circulaire (Pie Chart)** : Un diagramme circulaire est un graphique circulaire divisé en secteurs, dont chacun représente une proportion ou un pourcentage d'un ensemble. Chaque secteur (ou "part de tarte") montre la taille relative d'une catégorie par rapport au total. Les diagrammes circulaires sont souvent utilisés pour illustrer la répartition ou la composition d'un ensemble de données.

## Exercice 2: Fonctions Matplotlib pour Différents Diagrammes

**Les principales fonctions de Matplotlib utilisées pour créer les types de diagrammes suivants:**
1. Line Chart: `plt.plot()`
2. Histogramme: `plt.hist()`
3. Diagramme en Barres: `plt.bar()` pour des barres verticales et `plt.barh()` pour des barres horizontales
4. Diagramme de Dispersion: `plt.scatter()`
5. Diagramme Circulaire: `plt.pie()`

## Exercice 3: Création de Diagrammes avec les Données Fournies

**Utilisez les données fournies pour créer les diagrammes suivants:**
1. Un Line Chart montrant l'évolution du PIB par habitant d'un pays ou d'une région spécifique sur une période choisie.
2. Un Histogramme représentant la distribution du PIB par habitant pour tous les pays pour une année spécifique.
3. Un Diagramme en Barres pour comparer le PIB par habitant de plusieurs pays ou régions pour une année donnée.
4. Un Diagramme de Dispersion pour examiner la relation entre le PIB par habitant de deux régions différentes sur une période de temps.
5. Un Diagramme Circulaire pour montrer la répartition du PIB par habitant entre différents pays ou régions pour une année spécifique.

### 1- Line chart

````python
us_data = data[data['Country Name'] == 'United States'].drop(['Country Name', 'Indicator Name', 'Indicator Code', 'Country Code'], axis=1).T
us_data.columns = ['GDP per Capita']
plt.figure(figsize=(10, 6))
us_data.plot(kind='line', legend=False)
plt.title('Evolution of GDP per Capita (United States, 2010-2020)')
plt.xlabel('Year')
plt.ylabel('GDP per Capita (US$)')
plt.show()
````

### 2- Histogramme pour tous les pays (2020)

````python
data_2020 = data['2020'].dropna()
plt.figure(figsize=(10, 6))
plt.hist(data_2020, bins=20, color='blue', alpha=0.7, edgecolor='black')
plt.title('Distribution of GDP per Capita (All Countries, 2020)')
plt.xlabel('GDP per Capita (US$)')
plt.ylabel('Number of Countries')
plt.show()
````

### 3- Diagramme en barre

````python
data_2020_selected = data[['Country Name', '2020']].set_index('Country Name').head(10)
plt.figure(figsize=(16, 12))
data_2020_selected.plot(kind='bar', legend=False)
plt.title('GDP per Capita Comparison (2020)')
plt.xlabel('Country')
plt.ylabel('GDP per Capita (US$)')
plt.show()
````

### 4-  Diagramme de Dispersion pour les États-Unis vs Chine (2010-2020)

````python
us_china_data = data[data['Country Name'].isin(['United States', 'China'])].drop(['Country Name', 'Indicator Name', 'Indicator Code', 'Country Code'], axis=1)
us_gdp = us_china_data.iloc[0]
china_gdp = us_china_data.iloc[1]
plt.figure(figsize=(10, 6))
plt.scatter(us_gdp, china_gdp, alpha=0.7, c=np.arange(len(us_gdp)), cmap='viridis')
plt.title('GDP per Capita Relationship (United States vs China, 2010-2020)')
plt.xlabel('GDP per Capita (US$) - United States')
plt.ylabel('GDP per Capita (US$) - China')
plt.show()
````

### 5- Diagramme Circulaire pour les pays sélectionnés (2020)

````python
data_2020_pie = data_2020_selected['2020'].dropna().sort_values().head(5)
plt.figure(figsize=(10, 6))
plt.pie(data_2020_pie, labels=data_2020_pie.index, autopct='%1.1f%%', startangle=140)
plt.title('GDP per Capita Share (Selected Countries, 2020)')
plt.show()
````
