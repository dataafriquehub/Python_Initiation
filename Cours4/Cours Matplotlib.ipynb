- **Analyse des données**

- **Efficacité**

### 1.3. Graphique linéaire (Line Chart) avec `Matplotlib`

Un graphique linéaire est l'une des techniques de visualisation de données la plus simples et la plus utilisée pour représenter des données sur une période continue. Il est généralement utilisé pour montrer la tendance d'une mesure (ou d'une variable) dans le temps. Les valeurs des données sont représentées sous forme de points reliés par des segments de ligne.

Dans `Matplotlib`, nous pouvons créer un graphique linéaire en appelant la méthode native `.plot()`, qui est une commande polyvalente, et qui prend un nombre arbitraire d'arguments :

````python
#Génération de 100 Elts entre 0 et 10
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
#fig = plt.figure() # Pour sauvegarder le graphe

plt.plot(x, y1, '-')
plt.plot(x, y2, '--');
````

Présenté ainsi ce graphique ne sert à rien, car on ne sait pas ce qui représente.

#### Ajout de titres et d'étiquettes

Les graphiques doivent exprimer toute l'histoire. Pour exprimer cette histoire, il est nécessaire d'ajouter des étiquettes qui indiquent ce qui est représenté, où et ce que cela représente.

````python
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--')

plt.title('Sinus et Cosinus entre 1 et 10') # Ajouter un titre à la figure
plt.xlabel("DataAfrique Hub") # Ajout de l'étiquette pour l'axe horizontal
plt.ylabel("Y") # Ajout de l'étiquette pour l'axe vertical
plt.show()
````

Mettre une legend

````python
plt.plot(x, np.sin(x), '-', label = "Sinus")
plt.plot(x, np.cos(x), '--', label="Cosinus")

plt.title('Sinus et Cosinus entre 1 et 10') # Ajouter un titre à la figure
plt.xlabel("X entre 0 et 10") # Ajout de l'étiquette pour l'axe horizontal
plt.ylabel("Y") # Ajout de l'étiquette pour l'axe vertical
plt.legend() # Ajout d'une legende
#plt.legend(["Sinus", "Cosinus"])
plt.show()
````

#### On peut également créer un graphique linéaire avec un seul argument.

````python
y = np.random.rand(10)
plt.plot(y)
````

#### Mettre du style à son ghaphe.

La fonction plt.plot prend en charge de nombreux arguments pour styliser les lignes et les marqueurs :

* ```color ou c``` : Définit la couleur de la ligne (couleurs supportées)
* ```linestyle ou ls``` : Choisir entre une ligne pleine et une ligne en pointillés
* ```linewidth ou lw``` : Définir la largeur d'une ligne
* ```markersize ou ms``` : Définir la taille des marqueurs
* ```markeredgecolor ou mec``` : Définit la couleur des bords des marqueurs
* ```markeredgewidth ou mew``` : Définit la largeur du bord des marqueurs
* ```markerfacecolor ou mfc``` : Définit la couleur de remplissage des marqueurs
* ```alpha``` : Opacité du tracé

````python
plt.plot(y, linestyle='--', marker='o', color='r'); #expliquer le ;
````

#### Un raccourci pour mettre du style

````python
plt.plot(y, '--ro') # r--o ro-- ro(scatter) r-o
#fmt = '[marker][line][color]'
````

Il existe de nombreuses autres options. Vous pouvez essayer ce qui suit :


## 1.4 Tracer des courbes consécutives avec `Matplotlib`

````python
plt.plot(y, "bo-")
plt.show()
plt.plot(y, "rx-")
plt.show()
````

````python
years = range(2000, 2012)
apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931, 0.934, 0.936, 0.937, 0.9375, 0.9372, 0.939]
oranges = [0.962, 0.941, 0.930, 0.923, 0.918, 0.908, 0.907, 0.904, 0.901, 0.898, 0.9, 0.896] #production par hectare

plt.plot(years, apples, marker='s', c='b', ls='-', lw=2, ms=8, mew=2, mec='navy')
plt.plot(years, oranges, marker='o', c='r', ls='--', lw=3, ms=10, alpha=.5)

plt.xlabel('Year')
plt.ylabel('Rendement (tonnes par hectare)')

plt.title("Rendements des cultures à Mbour")
plt.legend(['Apples', 'Oranges']);
````

### 1.5. Plusieurs graphiques dans une seule figure en utilisant `Matplotlib`.
Nous pouvons créer plusieurs graphiques dans une seule figure. Ceci est très utile pour comparer des graphiques ou pour partager facilement des données provenant de plusieurs types de graphiques avec une seule image.

La méthode `.subplot()` est utilisée pour ajouter plusieurs graphiques dans une figure. Elle prend plusieurs arguments.

````python
#Rappel Figisize
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10,6)) #creating a grid of 2 rows, 2 columns and 6x6 figure size
ax[0,0].plot(years, apples, "go") # The top-left axes
ax[0,1].plot(years, apples, "bo") # The top-right axes
ax[1,0].plot(years, apples, "yo") # The bottom-left axes
ax[1,1].plot(years, apples, "ro") # The bottom-right axes

plt.show()
````

# Rappel

````python
# Graphique Linéaire
pib_par_habitant_afe = data[data['Country Code'] == 'AFE'].iloc[:, 4:-2].transpose()
#print()
pib_par_habitant_afe.columns = ['PIB par habitant']
pib_par_habitant_afe['Year'] = pib_par_habitant_afe.index
pib_par_habitant_afe['Year'] = pd.to_numeric(pib_par_habitant_afe['Year'])
pib_par_habitant_afe.head()
````

````python
plt.figure(figsize=(10, 6))
plt.plot(pib_par_habitant_afe['Year'], pib_par_habitant_afe['PIB par habitant'])
plt.title('Évolution du PIB par habitant en Afrique de l\'Est et du Sud')
plt.xlabel('Année')
plt.ylabel('PIB par habitant (USD)')
plt.grid(True)
plt.show()
````

## 1.6 Histogramme

Un *histogramme* représente la distribution d'une variable en créant des cases (intervalles) le long de la plage de valeurs et en affichant des barres verticales pour indiquer le nombre d'observations dans chaque case.
[En savoir plus...](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html)

````python
latest_year = data.columns[-3]
pib_distribution = data[latest_year].dropna()
````

````python
plt.figure(figsize=(10, 6))

plt.hist(pib_distribution, bins=30, edgecolor='black')

plt.title(f'Distribution du PIB par habitant dans le monde pour l\'année {latest_year}')
plt.xlabel('PIB par habitant (USD)')
plt.ylabel('Nombre de Pays')
plt.show();
````

## 1.7 Diagramme en Barre

Les diagrammes à barres sont assez semblables aux diagrammes linéaires, c'est-à-dire qu'ils montrent une séquence de valeurs. Cependant, chaque valeur est représentée par une barre et non par des points reliés par des lignes.

````python
countries_of_interest = ['NGA', 'KEN', 'GHA', 'EGY', 'ZAF'] # Nigeria, Kenya, Ghana, Égypte, Afrique du Sud
pib_par_habitant_selected = data[data['Country Code'].isin(countries_of_interest)][['Country Name', latest_year]]
pib_par_habitant_selected
````

````python
plt.figure(figsize=(10, 6))

plt.bar(pib_par_habitant_selected['Country Name'], pib_par_habitant_selected[latest_year])

plt.title(f'PIB par habitant en {latest_year} pour certains pays africains')
plt.xlabel('Pays')
plt.ylabel(f'PIB par habitant en {latest_year} (USD)')
plt.show()
````

````python
years = range(2000, 2006)
apples = [0.35, 0.6, 0.9, 0.8, 0.65, 0.8]
oranges = [0.4, 0.8, 0.9, 0.7, 0.6, 0.8]
plt.bar(years, apples)
plt.bar(years, oranges, bottom=apples);
````

## 1.8 Diagramme de dispersion (Scatter Plot)

Essayons de visualiser la relation entre la longueur et la largeur des sépales. Notre premier réflexe serait de créer un graphique linéaire à l'aide de plt.plot.

````python
plt.plot(data['2000'], data['2010']);
````

* Avec Matplotlib on utilise la fonction scatter() pour créer un scatter plot. [En savoir plus...](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html)

````python
colors = np.random.rand(len(data['2000']))
plt.figure(figsize=(10, 6))

plt.scatter(data['2000'], data['2010'], alpha=0.5, c=colors)

plt.title('Corrélation entre le PIB par habitant en 2000 et 2010')
plt.xlabel('PIB par habitant en 2000 (USD)')
plt.ylabel('PIB par habitant en 2010 (USD)')
plt.grid(True)
plt.show()
````

Utilisons Seaborn pour aavoir plus d'informations.

Un peu plus de style avec Matplotlib

````python
size = (30 * np.random.rand(len(data['2000'])))**2
colors = np.random.rand(len(data['2000']))

plt.figure(figsize=(10, 6))
plt.scatter(data['2000'], data['2010'], s=size, alpha=0.5, c=colors)

plt.title('Corrélation entre le PIB par habitant en 2000 et 2010')
plt.xlabel('PIB par habitant en 2000 (USD)')
plt.ylabel('PIB par habitant en 2010 (USD)')
plt.grid(True)
plt.show()
````

## 1.9 Diagramme circulaire

Un diagramme circulaire présente les données sous la forme d'un graphique circulaire divisé en tranches, chaque tranche représentant une proportion ou un pourcentage de l'ensemble.

````python
# Sélectionner l'année la plus récente disponible dans les données
latest_year = data.columns[-3]

# Sélectionner des pays spécifiques pour le diagramme circulaire
countries_for_pie = ['NGA', 'KEN', 'GHA', 'EGY', 'ZAF']  # Nigeria, Kenya, Ghana, Égypte, Afrique du Sud
pie_data = data[data['Country Code'].isin(countries_for_pie)][['Country Name', latest_year]]

# Préparer les données pour le diagramme circulaire
pie_labels = pie_data['Country Name']
pie_values = pie_data[latest_year]
````

````python
# Créer le diagramme circulaire
plt.figure(figsize=(6, 6))

plt.pie(pie_values, labels=pie_labels, autopct='%1.1f%%', startangle=140)

plt.title(f'Part du PIB par habitant en 2021 pour certains pays africains')
plt.show()
````
